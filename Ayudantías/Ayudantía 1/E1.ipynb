```python
class Vehiculo(): #Clase padre
  def __init__(self, nombre):
    self.nombre = nombre

  def mover(self):
    pass # Método sin implementar

  def __str__(self):
    return f"Vehículo: {self.nombre}" # Método str

class Motor: # Clase que pertenece como atributo de otra
    def __init__(self, tipo):
        self.tipo = tipo  # Puede ser gasolina, eléctrico, diésel, etc.

class Automovil(Vehiculo): # Subclase
  def __init__(self, nombre, tipo_motor):
    super().__init__(nombre)
    self.tipo_motor = tipo_motor

  def mover(self):
     return f"El automóvil {self.nombre} se mueve con un motor {self.motor.tipo}." # Overriding

class Bicicleta(Vehiculo): # Subclase
  def __init__(self, nombre, tipo_bici):
    super().__init__(nombre)
    self.tipo_bici = tipo_bici

  def mover(self):
    return f"La bicicleta {self.nombre} se mueve con pedales."
```

```python
# Crear objetos
auto1 = Automovil("Toyota Corolla", "Gasolina")
auto2 = Automovil("Tesla Model 3", "Eléctrico")
bici1 = Bicicleta("Bicicleta Montaña", "Montaña")
bici2 = Bicicleta("Bicicleta Urbana", "Urbana")

# Almacenar en una lista
lista_vehiculos = [auto1, auto2, bici1, bici2]

# Almacenar en un diccionario
diccionario_vehiculos  = {
    "auto1": auto1,
    "auto2": auto2,
    "bici1": bici1,
    "bici2": bici2
}

# Mostrar la lista
print("Lista de vehículos:")
for vehiculo in lista_vehiculos:
    print(f"- {vehiculo}")

# Mostrar el diccionario
print("\nDiccionario de vehículos:")
for nombre, vehiculo in diccionario_vehiculos.items():
    print(f"- {nombre}: {vehiculo}")
```
